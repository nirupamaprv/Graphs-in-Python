```python
# libraries
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
 
# Build a dataframe with 4 connections
df = pd.DataFrame({ 'from':['A', 'B', 'C','A'], 'to':['D', 'A', 'E','C']})
df
 
# Build your graph
G=nx.from_pandas_dataframe(df, 'from', 'to')
 
# Plot it
nx.draw_networkx(G, with_labels=True)
plt.show()
```

```python
df1 = pd.DataFrame({'start':['A','B','C'], 'end':['B','C','A']})
df1

G1 = nx.from_pandas_dataframe(df1,'start','end')

nx.draw_networkx(G1, with_labels=True)
plt.show()
```

```python
df2 = pd.DataFrame({'start':['A','B','C','E'], 'end':['B','C','A', 'E']})
df2

G1 = nx.from_pandas_dataframe(df2,'start','end')

nx.draw_networkx(G1, with_labels=True)
plt.show()
```

```python
#create an empty graph 
Gx = nx.Graph()
```

```python
#Add nodes
Gx.add_node(1)
```

```python
#Add Multiple nodes
Gx.add_nodes_from([2, 3, 4, 5])
```

```python
#Add edges
Gx.add_edge(1, 2)
```

```python
#Add edges by unpacking tuple
e = (2, 3)
Gx.add_edge(*e)  # unpack edge tuple*
```

```python
# adding a list of edges
Gx.add_edges_from([(4, 2), (1, 3), (5,4), (2,3)])
```

```python
#Plot abov created graph
nx.draw_networkx(Gx, with_labels=True)
plt.show()
```

```python
#Change color of nodes in plot
# Using parameter node_color as blue
nx.draw_networkx(Gx, node_color='b', with_labels=True)
plt.show()
```
